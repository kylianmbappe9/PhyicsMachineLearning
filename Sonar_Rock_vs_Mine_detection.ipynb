here i divided into X and Y where column X i have only included column number from 0 to 58 excluding 59 where column number 59 I have included in the Y column or Y axis

```python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

dividing the sets of datas into training data and testing data

I have split the datasets for X axis and Y axis into X_train, X_test, Y_train, Y_test and using the train_test split I have stratified two different sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

got the dimensitons for X axis, the X_train axis and X_test axis

```python
print(X.shape, X_train.shape, X_test.shape)
```

got the dimensitons for Y axis, the Y_train axis and Y_test axis

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
print(X_train)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```
